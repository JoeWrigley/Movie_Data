```python
#import code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

import seaborn as sns
```

```python
#import data (CSV file)
df = pd.read_csv('movies.csv')
```

```python
# Look at the Data
df.head()
```

```python
# Check for missiong data (for loop)

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
#Drop missing values

new_data = df.dropna()
new_data
for col in new_data.columns:
    pct_missing = np.mean(new_data[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    

```

```python
# Reassign back to df
df = new_data
df.head()
```

```python
df.dtypes
```

```python
#Change Colums Types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
# The year colomn and the (year) in the released column are not always matching. 
# Therefore, a new column was created to take the year in the released and make that the year

#Step 1: Change data type from object to string
df['year_correct'] = df['released'].astype(str).str[0:4]
#Drop the uwanted year column
df.drop(columns=['year'])
```

```python
#Order by Gross
highest_gross = df.sort_values(by=['gross'], inplace=False, ascending=False)
highest_gross.head(10)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
plt.show()

# plot the budget vs gross using seaborn


```

```python
#Correlation [Person] (numerical fields only)
df.corr()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df.nums = df

for col_name in df.nums.columns:
    if(df.nums[col_name]).dtype == 'object':
        df.nums[col_name] = df.nums[col_name].astype('category')
        df.nums[col_name] = df.nums[col_name].cat.codes

df.nums
```

```python
corr_matrix = df.nums.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
mat_unstack = df.nums.corr()

corr_pairs = mat_unstack.unstack()

corr_pairs.astype('float')

high_corr = corr_pairs[(corr_pairs)>0.6]
high_corr




```
